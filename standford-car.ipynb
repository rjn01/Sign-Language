```
from keras.preprocessing.image import ImageDataGenerator

train_gen = ImageDataGenerator(rescale=1/255.0, horizontal_flip = True, zoom_range = 0.2, shear_range=0.2)

train = train_gen.flow_from_directory(
        "/content/car_data/car_data/train",
        target_size = (256,256),
        class_mode="categorical",
        shuffle = True,
        batch_size=32)

test_gen = ImageDataGenerator(rescale=1/255.0 , validation_split=0.5)

validation = test_gen.flow_from_directory(
    "/content/car_data/car_data/test",
    target_size=(256,256),
    subset="training",
    shuffle = True,
    batch_size=32
)

test = test_gen.flow_from_directory(
    "/content/car_data/car_data/test",
    target_size=(256,256),
    subset = "validation"  )
```

# **MODEL**

```
from keras.applications.mobilenet import MobileNet
from keras.models import Model
from keras.applications.resnet_v2 import ResNet152V2
from keras.applications.densenet import DenseNet121


#from keras.applications.vgg16 import VGG16
```

```
model = DenseNet121(include_top = False, weights = "imagenet", input_shape=(256,256,3),classes=196)
model.summary()
```

```
for i in range(len(model.layers)):
  print(i,model.layers[i])
  model.layers[i].trainable=False
```

```
model.summary()
```

```
from keras.layers import *
```

```
x = Conv2D(128,(3,3),activation="relu")(model.output)
y = BatchNormalization()(x)
l1 = GlobalAveragePooling2D()(y)
l2 = Dropout(0.4)(l1)
l7 = Dense(196,activation="softmax")(l2)
model_new = Model(inputs=model.input,outputs=l7)
model_new.summary()
```

```
l = Conv2D(256,(3,3),activation="relu")(model.output)
l0 = BatchNormalization()(l)
a = MaxPooling2D()(l0)
#l = Conv2D(128,(3,3),activation="relu")(a)
#l0 = BatchNormalization()(l)
la = Conv2D(128,(3,3),activation="relu")(l0)
lb = BatchNormalization()(la)
l1 = GlobalAveragePooling2D()(lb)
l2 = Dropout(0.4)(l1)
'''
l3 = Dense(1024,activation="relu")(l2)
l4 = Dropout(0.4)(l3)
l5 = Dense(256,activation="relu")(l4)
l6 = Dropout(0.4)(l5)'''
l7 = Dense(196,activation="softmax")(l2)
model_new = Model(inputs=model.input,outputs=l7)
model_new.summary()
```

```
from keras.optimizers import SGD
from keras.optimizers import Adam
```

```
adam = Adam(lr=0.009)
sgd = SGD(lr=0.001, decay=1e-6, momentum=0.9)
model_new.compile(optimizer=sgd, loss="categorical_crossentropy",metrics=['accuracy'])
```

```
from keras.callbacks import ModelCheckpoint,EarlyStopping
```

```
import numpy as np

steps = int(12041/64)
steps
```

# **TRAINING THE MODEL**

```
filepath="/content/drive/My Drive/PROJECTS/CARS/best.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=True, mode='min')
callback = EarlyStopping(monitor='val_loss', patience=3)
call = [callback]
his = model_new.fit(train,epochs=75,validation_data=validation,steps_per_epoch=steps)
```

```
import matplotlib.pyplot as plt
plt.style.use("seaborn")
plt.plot(his.history["accuracy"],label="accuracy")
plt.plot(his.history["val_accuracy"],label="val_accuracy")
plt.xlabel("epochs")
plt.ylabel("percentage")
plt.title("acc % vs epochs")
plt.legend()
plt.show()
```

```
import matplotlib.pyplot as plt
plt.style.use("seaborn")
plt.plot(his.history["loss"],label="loss")
plt.plot(his.history["val_loss"],label="val_loss")
plt.xlabel("epochs")
plt.ylabel("percentage")
plt.title("loss % vs epochs")
plt.legend()
plt.show()
```
